```python
### Uzyskiwanie informacji z tensorów
# Pracując z tensorami trzeba być swiadomym ich atrybutów 
# Shape - długość(liczba elementów każdego wymiaru tensora) - tensor.shape
# Rank - liczba wymiarów (scalar ma rank 0, vector ma rank 1, matrix ma rank 2, tensor ma rank n) tensor.ndim
# Axis of dimension - wybrany wymiar tensora - tensor[0], tensor[:, 1]...
# Size - łączna ilość przedmiotów w tensorze - tf.size(tensor)
```

```python
# Stwórz tensor rank 4
rank_4_tensor = tf.zeros(shape=[2, 3, 4, 5])
rank_4_tensor
```

```python
rank_4_tensor[0]
```

```python
rank_4_tensor.shape, rank_4_tensor.ndim, tf.size(rank_4_tensor)
```

```python
# Zbierz poszczególne atrybuty tensora
print("Typ danych każdego elementu: ", rank_4_tensor.dtype)
print("Ilość wybiarów(rank): ", rank_4_tensor.ndim)
print("Krztałt tensora: ", rank_4_tensor.shape)
print("Elementy osi 0: ", rank_4_tensor.shape[0])
print("Elementy ostatniej osi: ", rank_4_tensor.shape[-1])
print("Łączna ilość elementów w tensorze: ", tf.size(rank_4_tensor))
print("Łączna ilość elementów w tensorze: ", tf.size(rank_4_tensor).numpy())
```

```python
# Wykonaj mnożenie macierzy pomiędzy X i Y (reshape)
tf.matmul(X, tf.reshape(Y, shape=(2, 3)))
```

```python
### Znajdź pozycje maximum i minimum tensora
```

```python
# Stwórz nowy tensor aby znaleźć pozycyjne minimum i maximum
tf.random.set_seed(42)
F = tf.random.uniform(shape=[50])
F
```

```python
# Znajdź pozycje maximum
tf.argmax(F)
```

```python
# Zindexuj naszą największą wartość pozycji
F[tf.argmax(F)]
```

```python
# Znajdżź maxymalną wartość F
tf.reduce_max(F)
```

```python
# Sprawdź równość
F[tf.argmax(F)] == tf.reduce_max(F)
```

```python
# Zindexuj najmniejszą wartość pozycji
tf.argmin(F)
```

```python
# Znajdź minimalną używaną wartość F
F[tf.argmin(F)]
```
